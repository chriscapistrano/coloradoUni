```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataTwo.csv', encoding = 'ISO-8859-1')
```

```python
df.head(50)
```

```python
df.columns
```

```python
#group by state
grouped = df.groupby('state')
```

```python
#summary by state
#number of schools in each state
schoolCount = grouped['chronname'].nunique()
schoolCount
```

```python
#schools within each state
schoolName = grouped['chronname'].unique()
schoolName
```

```python
#means per state
schoolMeans = grouped['student_count', 'awards_per_value','med_sat_value', 'grad_100_value', 'cohort_size', 'vsa_enroll_after4_first', 'aid_value'].mean()
schoolMeans
```

```python
#national stats (ALL SCHOOLS IN THE NATION) <-- this can be used to compare Colorado stats. How do we fare nationwide?
nationMeans = schoolMeans.mean()
nationMeans
```

```python
#Colorado-specific stats [*****consider comparing Colorado to national average stats]
colorado = df.loc[df['state'] == 'Colorado']
colorado
```

```python
#unique list of schools in Colorado, etc
coloCount = colorado['chronname'].nunique()
coloCount
```

```python
coloSchools = colorado['chronname'].unique()
coloSchools
```

```python
#Colorado stats
coloStats = colorado[['student_count', 'awards_per_value','med_sat_value', 'grad_100_value', 'cohort_size', 'vsa_enroll_after4_first', 'aid_value']].mean()
coloStats
```
